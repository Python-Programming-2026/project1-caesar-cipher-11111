```python
a=input()
alist=a.split(" ",1)
key=int(alist[0])
stxt=alist[1]
dtxt=""
for x in stxt:
    y=ord(x)
    if "a"<=x<="z":
        dn=(y-97+key)%26+97
    elif "A"<=x<="Z":
        dn=(y-65+key)%26+65
    else:
        dn=y
    dtxt=dtxt+chr(dn)
print(dtxt)
```

```python
def caesar_encrypt(plaintext,shift):
    """
    使用凯撒密码加密明文
    plaintext: 明文字符串
    shift: 移位量
    """
    result = []
    for ch in plaintext:
        if ch.isupper():
            result.append(chr((ord(ch)-ord('A')+shift)%26+ord('A')))
        elif ch.islower():
            result.append(chr((ord(ch)-ord('a')+shift)%26+ord('a')))
        else:
            result.append(ch)
    return ''.join(result)


def caesar_decrypt(ciphertext, shift):
    """
    使用凯撒密码解密密文
    ciphertext: 密文字符串
    shift: 加密时使用的移位量
    """
    return caesar_encrypt(ciphertext,26-(shift%26))

def main():
    while True:
        print("\n请选择操作：")
        print("1.加密")
        print("2.解密")
        print("3.退出")
        choice=input("输入选择（1/2/3）：").strip()

        if choice=='3':
            break
        elif choice in ('1','2'):
            text=input("请输入文本：")
            try:
                shift=int(input("请输入移位量（整数）："))
            except ValueError:
                print("移位量必须为整数，请重新输入。")
                continue

            if choice=='1':
                result=caesar_encrypt(text,shift)
                print("加密结果：",result)
            else:
                result=caesar_decrypt(text, shift)
                print("解密结果：",result)
        else:
            print("无效选择，请重新输入。")
            
if __name__ == "__main__":
    main()
```
