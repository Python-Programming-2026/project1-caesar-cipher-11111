```python
import tkinter as tk
from tkinter import messagebox

def caesar_encrypt(plaintext,shift):
    """
    使用凯撒密码加密明文
    plaintext: 明文字符串
    shift: 移位量
    """
    result=[]
    for ch in plaintext:
        if ch.isupper():
            result.append(chr((ord(ch)-ord('A')+shift)%26+ord('A')))
        elif ch.islower():
            result.append(chr((ord(ch)-ord('a')+shift)%26+ord('a')))
        else:
            result.append(ch)
    return ''.join(result)

def caesar_decrypt(ciphertext, shift):
    """
    使用凯撒密码解密密文
    ciphertext: 密文字符串
    shift: 加密时使用的移位量
    """
    return caesar_encrypt(ciphertext,26-(shift%26))

def process():
    # 获取移位量并转化为整数
    try:
        shift=int(shift_entry.get())
    except ValueError:
        messagebox.showerror("错误","移位量必须为整数")
        return

    # 获取输入文本
    text=input_text.get("1.0",tk.END)
    if not text:
        messagebox.showwarning("警告","请输入文本")
        return

    # 选择加密或者解密
    if mode_var.get()=="encrypt":
        result=caesar_encrypt(text,shift)
    else:
        result=caesar_decrypt(text,shift)

    # 显示结果
    output_text.delete("1.0",tk.END)
    output_text.insert("1.0",result)

# 创建主窗口
root = tk.Tk()
root.title("凯撒密码")
root.geometry("500x400")

# 输入标签和文本框
tk.Label(root, text="输入文本：").pack(pady=5)
input_text = tk.Text(root, height=5, width=50)
input_text.pack(pady=5)

# 移位量输入
tk.Label(root, text="移位量（整数）：").pack(pady=5)
shift_entry = tk.Entry(root, width=10)
shift_entry.pack(pady=5)

# 加密/解密选择
mode_var = tk.StringVar(value="encrypt")
mode_frame = tk.Frame(root)
mode_frame.pack(pady=5)
tk.Radiobutton(mode_frame, text="加密", variable=mode_var, value="encrypt").pack(side=tk.LEFT, padx=10)
tk.Radiobutton(mode_frame, text="解密", variable=mode_var, value="decrypt").pack(side=tk.LEFT, padx=10)

# 执行按钮
tk.Button(root, text="执行", command=process, bg="lightblue", width=15).pack(pady=10)

# 结果显示
tk.Label(root, text="结果：").pack(pady=5)
output_text = tk.Text(root, height=5, width=50)
output_text.pack(pady=5)

# 运行主循环
root.mainloop()
```
